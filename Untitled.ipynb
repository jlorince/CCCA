```python
%matplotlib inline
import pandas as pd
import json
import numpy as np

episode_json = json.loads('[%s]' % ','.join(open('data/episode-metadata_the-colbert-report').read().strip().split('\n')))
ep_data_colbert = pd.read_json(json.dumps(episode_json))
episode_json = json.loads('[%s]' % ','.join(open('data/episode-metadata_the-daily-show-with-jon-stewart').read().strip().split('\n')))
ep_data_stewart = pd.read_json(json.dumps(episode_json))
ep_data_colbert['show'] = 'colbert'
ep_data_stewart['show'] = 'stewart'
ep_data = pd.concat([ep_data_colbert,ep_data_stewart]).reset_index()

clip_json = json.loads('[%s]' % ','.join([line[line.find('{'):] for line in open('data/clip-metadata_the-colbert-report').read().strip().split('\n')]))
clip_data_colbert = pd.read_json(json.dumps(clip_json))
clip_json = json.loads('[%s]' % ','.join([line[line.find('{'):] for line in open('data/clip-metadata_the-daily-show-with-jon-stewart').read().strip().split('\n')]))
clip_data_stewart = pd.read_json(json.dumps(clip_json))
clip_data_colbert['show'] = 'colbert'
clip_data_stewart['show'] = 'stewart'
clip_data = pd.concat([clip_data_colbert,clip_data_stewart]).reset_index()
```

```python
def get_id(season):
    if season: return season.get('episodeNumber')
    else: return None
ep_data['ep_id'] = ep_data['season'].apply(get_id)
```

```python
def get_tags(tags):
    if type(tags)==list:
        return [tag['title'] for tag in tags]
    else:
        return tags
clip_data['tags'] = clip_data['tags'].apply(get_tags)
clip_data['date'] = pd.to_datetime(clip_data['airDate'],unit='s')


```

```python
x = clip_data[['date','show','tags','shortId']].set_index('date').tz_localize('UTC').tz_convert('US/Eastern')
```

```python
#x.groupby(by=(x.index['date'].year,x.index['date'].month))
from matplotlib import pyplot as plt
from collections import Counter
def grouper(grp):
    out = []
    for i,row in grp.iterrows():
        out+=row['tags']
    return pd.DataFrame({'tags':Counter(out)})
test = x.dropna().groupby([pd.TimeGrouper(freq='M'),'show']).apply(grouper)
datetime_idx = test.index.levels[0]
test = test.reset_index()
test.columns = ['date','show','tag','freq']
```

```python
for col,show in zip(('blue','red'),('stewart','colbert')):
    fig,axes = plt.subplots(4,4,figsize = (16,16))
    rav = axes.ravel()
    top_tags = test[test['show']==show]['tag'].value_counts()
    for i,tag in enumerate(top_tags.index[:16]):
        ax = rav[i]
        current = test[(test['show']==show) & (test['tag']==tag)].set_index('date').reindex(datetime_idx).fillna(0)['freq']
        current.plot(kind='line',ax=ax,rot=90,c=col)
        ax.set_title(tag)
```
